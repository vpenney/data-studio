```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import NullFormatter
from matplotlib.ticker import FuncFormatter
import matplotlib.cm as cm

%matplotlib inline
import numpy as np
from cycler import cycler
import pandas as pd
import re
```

```python
df = pd.read_csv('beer_by_state.csv')
df
```

```python
df['avg'] = df['barrels_produced']/ df['total_breweries']
```

```python
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
ax = df.plot(kind='scatter', x='total_breweries', y='barrels_produced', figsize=(15,10), s=40)
# CUSTOMIZATION CODE GOES UNDER HERE
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'patch.linewidth': 0.5,
    'figure.facecolor': '#C7DCEB',
    'font.size': 11.0,
    'axes.facecolor': '#C7DCEB',
    'axes.edgecolor': 'white',
    'axes.linewidth': '1',
    'axes.grid': True,
    'grid.color': 'white',
    'axes.titlesize': 'x-large',
    'axes.labelsize': 'medium',
    'axes.axisbelow': True,
    'axes.prop_cycle': cycler('color', ['#203E7D', '#5286FF', '#3DE894', '#F4FF90', '#E8AB55', '#FF708B']),
    'xtick.major.size': 0,
    'xtick.minor.size': 0,
    'ytick.major.size': 0,
    'ytick.minor.size': 0,
    'axes.labelweight': 'medium',
    'axes.spines.left': False,
    'axes.spines.bottom': False,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'legend.facecolor': '#FFFFFF',
    'font.family': 'Franklin Gothic Book',
    'lines.markersize': 10
})


# plt.show()
```

```python
# Step One: build a function
def with_commas(value, pos): 
    millions = value / 100000
    return "{:,}".format(int(value))

# Step Two: Use FuncFormatter with that function
formatter = FuncFormatter(with_commas)

fig, ax = plt.subplots()
ax = df.plot(kind='scatter', x='total_breweries', y='barrels_produced', figsize=(15,10), s=90, c='avg', cmap='Spectral_r', alpha=.9, edgecolors='#1B3D7F')
ax.set_title("Craft beer production by state", pad=20)
ax.set_xlabel("Number of breweries")
ax.set_ylabel("Barrels of beer produced")
ax.title.set_fontsize(16)
ax.grid(False, axis='x')
ax.yaxis.set_major_formatter(formatter)

plt.savefig("production_vs_number.pdf", facecolor=plt.gcf().get_facecolor(), bbox_inches=[])
plt.show()
```

```python
ax = df.sort_values(by=['avg'], ascending=False).tail(5).plot(kind='barh', x='state', y='avg', legend=False)
ax.legend=False
ax.set_title("States with least productive breweries (2017)")
ax.set_xlabel("Average barrel production per brewery")
ax.set_ylabel("")
ax.grid(False, axis='y')

plt.tight_layout()
plt.savefig("least-productive-states.pdf", facecolor=plt.gcf().get_facecolor())
```

```python
ax = df.sort_values(by=['avg'], ascending=True).tail(5).plot(kind='barh', x='state', y='avg', legend=False)
ax.legend=False
ax.set_title("States with most productive breweries (2017)")
ax.set_xlabel("Average barrel production per brewery")
ax.set_ylabel("")
ax.grid(False, axis='y')

plt.tight_layout()
plt.savefig("most-productive-states.pdf", facecolor=plt.gcf().get_facecolor())
```

```python
ax = df.sort_values(by=['breweries_per_capita']).plot(kind='barh', x='state', y='breweries_per_capita', color='#203E7D', figsize=(7,20), legend=False)
ax.set_title("Breweries per capita", pad=20)
ax.grid(False)
ax.set_xlabel("*Breweries per 100,000 adults 21+")
ax.set_ylabel('')
ax.title.set_fontsize(16)
ax.xaxis.set_major_formatter(NullFormatter())
ax.xaxis.set_minor_formatter(NullFormatter())
y = df['breweries_per_capita']


# label values inside of bars
for i, bar in enumerate(ax.patches):
    value = bar.get_width()
    if i==50:
        value = str(bar.get_width()) + " breweries"
    else: 
        value = bar.get_width()
    x_pos = bar.get_width() - .08
    y_pos = i
    ax.text(x_pos, y_pos, value, va='center', ha='right', color='white', fontweight='bold')


plt.tight_layout()
plt.savefig("breweries_per_capita.pdf", facecolor=plt.gcf().get_facecolor())
plt.show()
```

```python
# Step One: build a function
def with_commas(value, pos): 
    millions = value / 100000
    return "{:,}".format(int(value))

# Step Two: Use FuncFormatter with that function
formatter = FuncFormatter(with_commas)

ax = df.plot(kind='scatter', x='barrels_produced', y='excise_tax', figsize=(12, 7), s=75, alpha=.75)

ax.set_xlabel("Barrels of beer produced per year")
ax.set_ylabel("Excise tax (percent)")
ax.grid(False, axis='x')
ax.set_title("Lower state excise taxes suggest more beer", pad=20)
ax.title.set_fontsize(16)
ax.xaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.savefig("excise_tax.pdf", facecolor=plt.gcf().get_facecolor(), bbox_inches=[])
plt.show()
```

```python
ax = df.sort_values(by=['excise_tax'], ascending=False).tail(5).plot(kind='barh', x='state', y='excise_tax', legend=False)
ax.legend=False
ax.set_title("States with lowest beer excise taxes (2017)")
ax.set_xlabel("%Tax")
ax.set_ylabel("")
ax.grid(False, axis='y')

plt.tight_layout()
plt.savefig("lowest-state-beer-taxes.pdf", facecolor=plt.gcf().get_facecolor())
```
