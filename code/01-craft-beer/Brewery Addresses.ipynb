```python
from time import sleep
import requests
import pandas as pd
import re
```

```python
df = pd.read_csv('brewery_addresses.csv')
df.head()
```

```python
df.address[0]
```

```python
df['longitude'] = 0
df['latitude'] = 0
df.head()
```

```python
import urllib

for n, address in enumerate(df.address):
    address = address.replace("[", "").replace("]", "").replace("'", "").replace(",", "") 
    address = address.split("|")[0]
    address = address.split("Type:")[0]
    address = urllib.parse.quote(address)
    sleep(0.25)
 
    try:
      # Form the URL with the address in it
      url = "http://maps.googleapis.com/maps/api/geocode/json?sensor=false&address={0}".format(address)

      # Request the URL
      response = requests.get(url)

      # Dig deep into the JSON 
      # this will give us something like
      # {u'lat': 40.7135296, u'lng': -73.9856844}
      coords = response.json()['results'][0]['geometry']['location']
    
      # Update row that's being scraped
      df.loc[n, 'latitude'] = coords['lat']
      df.loc[n, 'longitude'] = coords['lng']
    
      # And now save it to the database
      print("{} is at {}, {}".format(df.name, df.latitude, df.longitude))
    except Exception as e: 
        print(e)
```
