```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
driver = webdriver.Chrome()
driver.get('https://www.brewersassociation.org/directories/breweries/?term=Alabama&searchby=statename')
```

```python
import time

# select the US
country_button = driver.find_element_by_id("country")
country_button.click()
time.sleep(3)
us = driver.find_element_by_xpath('//*[@id="country_select"]/li[1]/a')
us.click()
```

```python
# filter by microbreweries, brewpubs, and craft breweries
micro_button = driver.find_element_by_xpath('//*[@id="micro"]')
micro_button.click()

brewpub_button = driver.find_element_by_xpath('//*[@id="type_select"]/label[2]')
brewpub_button.click()

craft_button = driver.find_element_by_xpath('//*[@id="type_select"]/label[5]')
craft_button.click()
```

```python
# scrape addresses and zip codes of breweries
import re 


brewery_info = driver.find_elements_by_class_name('brewery')
rows = []

for brewery in brewery_info:
    breweries = {}
    full_info = brewery.text.split('\n')
    try:
        
        breweries['name'] = full_info[0]
        breweries['address'] = full_info[1:3]
    except:
        pass
    
    rows.append(breweries)
print(rows)
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv('brewery_addresses.csv', index=False)
```

```python
driver = webdriver.Chrome()
driver.get('https://www.brewersassociation.org/statistics/by-state/')
```

```python
# grab the state name, number of breweries, breweries per capita (100k humans 21+), and barrels produced annually
by_state = driver.find_elements_by_id('state-stats')
goodbeerstuff = []

for state in by_state:
    state_beer = {}
    name = state.find_element_by_id('state-header').text.split('\n')
    try:
        state_beer['state'] = name[0]
        info = state.text.split('\n')
        state_beer['total_breweries'] = info[3]
        state_beer['breweries_per_capita'] = info[6]
        state_beer['barrels_produced'] = info[17]
    except:
        pass
    goodbeerstuff.append(state_beer)
print(goodbeerstuff)
#print(info[1].text)
```

```python
re.findall(" Breweries per Capita*",goodbeerstuff)
```

```python
df_beerpeople = pd.DataFrame(goodbeerstuff)
df_beerpeople
```

```python
df_beerpeople.dtypes
df_beerpeople['total_breweries'] = df_beerpeople['total_breweries'].apply(pd.to_numeric)
```

```python
df_beerpeople['breweries_per_capita'] = df_beerpeople['breweries_per_capita'].str.replace(' Breweries per Capita\*', '')
df_beerpeople['breweries_per_capita'].apply(pd.to_numeric)
df_beerpeople['barrels_produced'] = df_beerpeople['barrels_produced'].str.replace(',', '')
df_beerpeople['barrels_produced'].apply(pd.to_numeric)
df_beerpeople.head()
```
