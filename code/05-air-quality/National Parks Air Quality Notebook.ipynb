```python
seasonal = pd.read_csv('epa_data/Concentration - Seasonal.csv')
seasonal
```

```python
rm_seasonal = seasonal[seasonal.SITE_ID.str.contains("ROM406")]
rm_seasonal['DATEON'] = pd.to_datetime(rm_seasonal['DATEON'], format='%m/%d/%Y %H:%M:%S')
```

```python
yel_seasonal = seasonal[seasonal.SITE_ID.str.contains("YEL")]
yel_seasonal['DATEON'] = pd.to_datetime(yel_seasonal['DATEON'], format='%m/%d/%Y %H:%M:%S')
```

```python
rm_seasonal.dtypes
```

```python
ax = rm_seasonal.sort_values(by='DATEON').plot(x='DATEON', y='TNO3_CONC', figsize=(11,6), kind='bar', color=['#275876', '#8bccc9', '#ec866c', '#f9d669', '#275876', '#8bccc9', '#ec866c', '#f9d669', '#275876', '#8bccc9', '#ec866c', '#f9d669', '#275876', '#8bccc9', '#ec866c', '#f9d669', '#275876', '#8bccc9', '#ec866c', '#f9d669',  '#275876', '#8bccc9'])
legend_elements = [Patch(facecolor='#275876', edgecolor='white',
                         label='Color Patch'),
                   Patch(facecolor='#8bccc9', edgecolor='white',
                         label='Color Patch'),
                   Patch(facecolor='#ec866c', edgecolor='white',
                         label='Color Patch'),
                   Patch(facecolor='#f9d669', edgecolor='white',
                         label='Color Patch')]
ax.legend(legend_elements, ['Winter', 'Spring', 'Summer', 'Fall'])
ax.set_xlabel('')
ax.set_title("Rocky Mtn. NP's Seasonal Air Quality")
ax.grid(False, axis='x')

plt.tight_layout()
plt.savefig("rm-air-quality.pdf")
plt.show()
```
