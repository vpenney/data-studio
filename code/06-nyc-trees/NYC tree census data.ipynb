```python
df1995 = pd.read_csv('1995_Street_Tree_Census.csv')
df2005 = pd.read_csv('2005_Street_Tree_Census.csv')
df2015 = pd.read_csv('2015_Street_Tree_Census_-_Tree_Data.csv')
```

```python
df2005.dtypes
```

```python
ax = df1995.groupby(['Borough']).size().sort_values().plot(kind='barh')
ax.grid(False, axis='y')
ax.set_title("Trees per borough, '95")

plt.show()
```

```python
ax = df2005.groupby(['boroname']).size().sort_values().plot(kind='barh')
ax.grid(False, axis='y')
ax.set_title("Trees per borough, '05")

plt.show()
```

```python
df2015.dtypes
```

```python
result = df2015.groupby(['borough']).size().sort_values(ascending=True)
ax = result.plot(kind='barh')
ax.grid(False, axis='y')
ax.set_title("Trees per borough, '15")

plt.show()
```
